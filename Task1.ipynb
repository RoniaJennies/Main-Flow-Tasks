```python
import pandas as pd

# Load dataset with proper separator
df = pd.read_csv("student-mat.csv", sep=';')

# Show first 5 rows
print(df.head())
```

```python
# Missing values
print("Missing values:\n", df.isnull().sum())
# Shape of the data
print("\nDataset shape:", df.shape)
```

```python
# Remove missing and duplicate data
df = df.dropna()
df = df.drop_duplicates()

print("Cleaned dataset shape:", df.shape)
```

```python
avg_g3 = df["G3"].mean()
print("Average Final Grade (G3):", avg_g3)
```

```python
count_above_15 = (df["G3"] > 15).sum()
print("Number of students with G3 > 15:", count_above_15)
```

```python
correlation = df["studytime"].corr(df["G3"])
print("Correlation between Study Time and G3:", correlation)
```

```python
gender_avg = df.groupby("sex")["G3"].mean()
print("Average G3 by gender:\n", gender_avg)
```

```python
import matplotlib.pyplot as plt

plt.hist(df["G3"], bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Final Grades (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")
plt.show()
```

```python
plt.scatter(df["studytime"], df["G3"], color='green')
plt.title("Study Time vs Final Grade (G3)")
plt.xlabel("Study Time (hours/week)")
plt.ylabel("Final Grade (G3)")
plt.show()
```

```python
gender_avg.plot(kind='bar', color=['orange', 'skyblue'])
plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average G3")
plt.xticks(rotation=0)
plt.show()
```
